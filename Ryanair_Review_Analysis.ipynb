```python
#! pip install gensim
#! pip install pyLDAvis

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

# read dataset
df_raw = pd.read_csv('ryanair_reviews.csv')

# choose the characteristics(remove unimportant columns)
keep_columns = ['Comment', 'Overall Rating', 'Seat Comfort', 'Cabin Staff Service', 'Value For Money', 'Recommended']
df = df_raw[keep_columns].copy()

# data cleaning
df = df.dropna(subset=['Comment'])
df = df[df['Comment'].str.strip() != ""]

# remove duplicate comments
df = df.drop_duplicates(subset=['Comment'])

print(f"cleaning finished！the remaining data entries: {len(df)}")
df.head()

import matplotlib.pyplot as plt

# calculate average overall rating
avg_rating = df['Overall Rating'].mean()
print(f"average overall rating: {avg_rating:.2f}")

fig, ax = plt.subplots()
df['Overall Rating'].value_counts().sort_index().plot(kind='bar', color='skyblue', ax=ax)

# average of overall rating line
ax.axvline(x=avg_rating - 1, color='red', linestyle='--', linewidth=1.5, label=f'Mean: {avg_rating:.2f}')

plt.title('Distribution of Overall Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.legend()
plt.show()

# Import pandas for data loading and data manipulation.
import pandas as pd

# Import NLTK for natural language processing tasks.
import nltk

# Import the VADER sentiment analyser from NLTK.
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon required for sentiment analysis.
nltk.download('vader_lexicon')

# Read the Ryanair review dataset from a CSV file using UTF-8 encoding.
df_raw = pd.read_csv('ryanair_reviews.csv', encoding='utf-8')

# Create a copy of the original dataset to avoid modifying the raw dataframe directly.
df = df_raw.copy()

# Define the column name that contains the passenger review text.
text_col = "Comment"

# Remove rows where the review text is missing.
df = df.dropna(subset=[text_col]).copy()

# Convert the review text column into string format to ensure consistent text processing.
df[text_col] = df[text_col].astype(str)

# Remove rows where the review text is empty after removing extra spaces.
df = df[df[text_col].str.strip() != ""].copy()

# Create a VADER sentiment analyser object.
sia = SentimentIntensityAnalyzer()

# Apply VADER sentiment analysis to each review and generate sentiment scores.
scores = df[text_col].apply(sia.polarity_scores)

# Convert the sentiment score results into a dataframe and add the prefix "vader_" to each score column.
scores_df = pd.DataFrame(list(scores)).add_prefix("vader_")

# Define the old VADER-related columns that may already exist from previous runs.
old_vader_cols = ["vader_neg", "vader_neu", "vader_pos", "vader_compound", "vader_sentiment"]

# Remove any existing VADER-related columns to avoid duplicated or outdated results.
df = df.drop(columns=[col for col in old_vader_cols if col in df.columns], errors="ignore")

# Combine the cleaned review dataset with the newly generated VADER sentiment score columns.
df = pd.concat([df.reset_index(drop=True), scores_df.reset_index(drop=True)], axis=1)

# Define a function to classify VADER compound scores into sentiment categories.
def classify_vader(compound_score):

    # Classify reviews with a compound score of 0.05 or above as positive.
    if compound_score >= 0.05:
        return "Positive"

    # Classify reviews with a compound score of -0.05 or below as negative.
    elif compound_score <= -0.05:
        return "Negative"

    # Classify reviews with compound scores between -0.05 and 0.05 as neutral.
    else:
        return "Neutral"

# Apply the VADER classification function to the compound score column.
df["vader_sentiment"] = df["vader_compound"].apply(classify_vader)


print(df[[text_col, "vader_neg", "vader_neu", "vader_pos", "vader_compound", "vader_sentiment"]].head(10))

# Count the number of reviews in each VADER sentiment category.
sentiment_counts = df["vader_sentiment"].value_counts()

# Calculate the percentage of reviews in each VADER sentiment category.
sentiment_percent = df["vader_sentiment"].value_counts(normalize=True) * 100

print("\nSentiment Counts:")
print(sentiment_counts)


print("\nSentiment Percentages:")
print(sentiment_percent.round(2))

# # Save the processed dataset with VADER sentiment scores and labels for record keeping and further analysis.
df.to_csv("Ryanair_VADER_results1.csv", index=False, encoding="utf-8-sig")
print("\nSaved: Ryanair_VADER_results1.csv")

# Import matplotlib for creating visualisations.
import matplotlib.pyplot as plt

# Import seaborn for creating more polished statistical charts.
import seaborn as sns

# Define the display order of sentiment categories in the bar chart.
sentiment_order = ["Positive", "Neutral", "Negative"]

# Calculate the percentage distribution of each VADER sentiment category.
sentiment_counts = (
    # Select the VADER sentiment classification column.
    df["vader_sentiment"]

    # Count the proportion of each sentiment category instead of raw counts.
    .value_counts(normalize=True)

    # Reorder the sentiment categories based on the predefined order.
    .reindex(sentiment_order)

    # Convert proportions into percentages.
    .mul(100)
)

# Create a figure with a specified size for the bar chart.
plt.figure(figsize=(7,5))

# Create a bar chart showing the percentage of each sentiment category.
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)

# Add a title to the bar chart.
plt.title("Distribution of VADER Sentiment Categories")

# Add a label to the x-axis.
plt.xlabel("Sentiment Category")

# Add a label to the y-axis.
plt.ylabel("Percentage (%)")

# Add percentage labels above each bar.
for i, v in enumerate(sentiment_counts.values):

    # Place the percentage value slightly above each bar.
    plt.text(i, v + 1, f"{v:.1f}%", ha='center')

# Set the y-axis limit to provide space for the percentage labels.
plt.ylim(0, max(sentiment_counts.values) + 8)

# Adjust the layout to prevent labels or titles from being cut off.
plt.tight_layout()
plt.show()

# Calculate the average overall rating for each VADER sentiment category.
overall_by_sentiment = (
    # Group the dataset by VADER sentiment category and select the overall rating column.
    df.groupby("vader_sentiment")["Overall Rating"]

    # Calculate the mean overall rating for each sentiment category.
    .mean()

    # Reorder the sentiment categories as Positive, Neutral, and Negative.
    .reindex(sentiment_order)
)

# Create a new figure and set its size.
plt.figure(figsize=(7,5))

# Create a bar chart showing the average overall rating for each sentiment category.
sns.barplot(x=overall_by_sentiment.index, y=overall_by_sentiment.values)
plt.title("Average Overall Rating by VADER Sentiment")
plt.xlabel("Sentiment Category")
plt.ylabel("Average Overall Rating")

# Add average rating labels above each bar.
for i, v in enumerate(overall_by_sentiment.values):

    # Place each average rating value slightly above its corresponding bar.
    plt.text(i, v + 0.1, f"{v:.2f}", ha='center')

# Set the y-axis range and leave extra space above the tallest bar.
plt.ylim(0, max(overall_by_sentiment.values) + 1)

# Adjust the layout to prevent chart elements from overlapping.
plt.tight_layout()
plt.show()

# Calculate the average value-for-money rating for each VADER sentiment category.
vfm_by_sentiment = (
    # Group the dataset by VADER sentiment category and select the Value For Money column.
    df.groupby("vader_sentiment")["Value For Money"]

    # Calculate the mean Value For Money score for each sentiment category.
    .mean()

    # Reorder the sentiment categories as Positive, Neutral, and Negative.
    .reindex(sentiment_order)
)

# Create a new figure and set its size.
plt.figure(figsize=(7,5))
# Create a bar chart showing the average Value For Money score for each sentiment category.
sns.barplot(x=vfm_by_sentiment.index, y=vfm_by_sentiment.values)
plt.title("Average Value for Money by VADER Sentiment")
plt.xlabel("Sentiment Category")
plt.ylabel("Average Value for Money")

# Add average Value For Money labels above each bar.
for i, v in enumerate(vfm_by_sentiment.values):
    # Place each average Value For Money score slightly above its corresponding bar.
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center')

# Set the y-axis range and leave extra space above the tallest bar.
plt.ylim(0, max(vfm_by_sentiment.values) + 0.5)
# Adjust the layout to prevent chart elements from overlapping.
plt.tight_layout()
plt.show()

# Identify reviews that were classified as positive by VADER but received a low overall rating.
mismatch_positive_low_rating = df[
    # Select reviews where the VADER sentiment label is positive.
    (df["vader_sentiment"] == "Positive") &
    # Select reviews where the overall rating is 3 or below.
    (df["Overall Rating"] <= 3)

# Keep only the columns needed for inspecting the mismatch cases.
][["Comment", "Overall Rating", "vader_compound", "vader_sentiment"]]

# Identify reviews that were classified as negative by VADER but received a high overall rating.
mismatch_negative_high_rating = df[
    # Select reviews where the VADER sentiment label is negative.
    (df["vader_sentiment"] == "Negative") &
    # Select reviews where the overall rating is 8 or above.
    (df["Overall Rating"] >= 8)

# Keep only the columns needed for inspecting the mismatch cases.
][["Comment", "Overall Rating", "vader_compound", "vader_sentiment"]]

# Loop through the first 10 mismatch cases where VADER classified the review as positive but the overall rating was low.
for i, row in mismatch_positive_low_rating.head(10).iterrows():
    print(f"Review Index: {i}")
    print(f"Overall Rating: {row['Overall Rating']}")
    print(f"VADER Compound: {row['vader_compound']}")
    print(f"Sentiment: {row['vader_sentiment']}")
    print("Comment:")
    print(row["Comment"])
    # Print a separator line to make each review easier to read.
    print("=" * 100)

# Loop through the first 10 mismatch cases where VADER classified the review as negative but the overall rating was high.
for i, row in mismatch_negative_high_rating.head(10).iterrows():
    print(f"Review Index: {i}")
    print(f"Overall Rating: {row['Overall Rating']}")
    print(f"VADER Compound: {row['vader_compound']}")
    print(f"Sentiment: {row['vader_sentiment']}")
    print("Comment:")
    print(row["Comment"])
    print("=" * 100)

# Import the regular expression module for text cleaning.
import re

# Import NLTK for natural language processing tasks.
import nltk

# Import the English stopword list from NLTK.
from nltk.corpus import stopwords

# Import the WordNet lemmatizer for reducing words to their base form.
from nltk.stem import WordNetLemmatizer

# Import the word tokenizer for splitting text into individual words.
from nltk.tokenize import word_tokenize

# Download the tokenizer resources required for word tokenisation.
nltk.download('punkt')
nltk.download('punkt_tab')

# Download the English stopword list.
nltk.download('stopwords')

# Download WordNet for lemmatisation.
nltk.download('wordnet')

# Download Open Multilingual WordNet to support WordNet lemmatisation.
nltk.download('omw-1.4')

# Load the standard English stopword list from NLTK.
stop_words = set(stopwords.words('english'))

# Create a WordNet lemmatizer object.
lemmatizer = WordNetLemmatizer()

# Combine standard English stopwords with dataset-specific stopwords.
custom_stopwords = stop_words.union({
    'flight', 'ryanair', 'fly', 'passenger', 'time', 'one',
    'airline', 'would', 'travel', 'go', 'said', 'us', 'also',
    'say', 'take', 'could', 'even', 'make', 'give', 'day',
    'first', 'two', 'told', 'plane', 'tell'
})

# Define a preprocessing function for preparing review text for LDA topic modelling.
def preprocess_for_lda(text):

    # Convert all review text to lowercase.
    text = text.lower()

    # Remove punctuation, numbers, and special characters, keeping only letters and spaces.
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Split the cleaned text into individual word tokens.
    tokens = word_tokenize(text)

    # Remove stopwords and words with fewer than three characters.
    tokens = [word for word in tokens if word not in custom_stopwords and len(word) > 2]

    # Apply basic lemmatisation to reduce words to their base form.
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Return the cleaned token list.
    return tokens

# Apply the preprocessing function to each review comment and store the token list in a new column.
df['lda_tokens'] = df['Comment'].apply(preprocess_for_lda)

# Check that the new lda_tokens column has been created correctly.
df[['Comment', 'lda_tokens']].head()

# Install gensim and pyLDAvis if they are not already installed.
# !pip install gensim pyLDAvis

# Import gensim for building the LDA topic model.
import gensim

# Import corpora from gensim to create the dictionary and corpus.
from gensim import corpora

# Import CoherenceModel to evaluate the quality of the topic model.
from gensim.models import CoherenceModel

# Import pyLDAvis for interactive topic model visualisation.
import pyLDAvis

# Import the gensim-specific pyLDAvis module.
import pyLDAvis.gensim_models

# Convert the preprocessed token lists into the input format required by gensim.
texts = df['lda_tokens'].tolist()

# Create a dictionary that maps each unique word to a numerical ID.
dictionary = corpora.Dictionary(texts)

# Remove very rare words and overly common words from the dictionary.
dictionary.filter_extremes(no_below=10, no_above=0.4)

# Convert each review into a bag-of-words representation using the dictionary.
corpus = [dictionary.doc2bow(text) for text in texts]

# Set the number of topics for the LDA model.
num_topics = 5

# Train the LDA topic model using the prepared corpus and dictionary.
lda_model = gensim.models.LdaModel(
    # Use the bag-of-words corpus as the input data.
    corpus=corpus,

    # Provide the dictionary that maps word IDs back to actual words.
    id2word=dictionary,

    # Set the model to generate five topics.
    num_topics=num_topics,

    # Set a random seed to make the model output reproducible.
    random_state=100,

    # Update the model after each training chunk.
    update_every=1,

    # Set the number of documents processed in each training chunk.
    chunksize=100,

    # Set the number of full passes through the corpus.
    passes=20,

    # Allow gensim to automatically learn the document-topic density parameter.
    alpha='auto',

    # Store per-word topic assignments for more detailed topic analysis.
    per_word_topics=True
)

# Create a coherence model to evaluate the interpretability of the five-topic LDA model.
coherence_model_lda = CoherenceModel(
    # Provide the trained LDA model.
    model=lda_model,

    # Provide the preprocessed token lists.
    texts=texts,

    # Provide the dictionary used to train the model.
    dictionary=dictionary,

    # Use the c_v coherence metric.
    coherence='c_v'
)

# Calculate the coherence score for the five-topic LDA model.
coherence_score = coherence_model_lda.get_coherence()

# Print the coherence score rounded to four decimal places.
print(f"5-topic coherence score: {coherence_score:.4f}")

# Print the keywords associated with each generated topic.
for idx, topic in lda_model.print_topics(-1):

    # Display the topic number and its most representative words.
    print(f"Topic: {idx} \nWords: {topic}\n")

# Enable pyLDAvis visualisation inside the notebook.
pyLDAvis.enable_notebook()

# Prepare the interactive LDA topic visualisation.
vis = pyLDAvis.gensim_models.prepare(
    # Provide the trained LDA model.
    lda_model,

    # Provide the bag-of-words corpus.
    corpus,

    # Provide the dictionary used by the model.
    dictionary,

    # Keep the original topic numbering instead of sorting topics by size.
    sort_topics=False
)

# Display the interactive topic visualisation.
vis
```
